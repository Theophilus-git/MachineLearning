```python
# Visualising the data using :
# SepalLength

sns.histplot(df['SepalLengthCm'],kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
# PetalLengthCm
sns.histplot(df['PetalLengthCm'],kde=True)
plt.title("Distribution of the Petal length")
plt.show()
```

```python
# PetalWidthCm

sns.histplot(df['PetalWidthCm'],kde=True)
plt.title("Distribution of the Petal Width")
plt.show()
```

```python
# A scatter plot of the SepalLengthCm

sns.scatterplot(x = df['SepalLengthCm'], y = df['SepalWidthCm'], hue=df['Species'])
plt.title("Sepal length vs Sepal Width")
plt.show()
```
